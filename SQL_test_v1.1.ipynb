```python
import sqlite3
import pandas as pd
```

# SETUP

```python
src = sqlite3.connect('test.db')
```

```python
invoice = pd.read_csv('data/invoice.csv', sep=';', decimal=',')
ratings = pd.read_csv('data/ratings.csv', sep=';', decimal=',')
payment_type = pd.read_csv('data/payment_type.csv', sep=';', decimal=',')
```

```python
invoice.to_sql('invoice', src, if_exists='replace', index=False)
ratings.to_sql('ratings', src, if_exists='replace', index=False)
payment_type.to_sql('payment_type', src, if_exists='replace', index=False)
```
